```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/hp/Downloads/indian_employee.csv')
```

```python
df.head(10)
```

## Data Cleaning

```python
#Checking the missing values
```

```python
print("Missing Values in each column")
df.isnull().sum()
```

```python
#Filling NaN values with a calucalated average value
```

```python
df['Salary (INR)'] = df['Salary (INR)'].fillna(df['Salary (INR)'].mean())
```

```python
df
```

```python
df['Experience (Years)'] = df['Experience (Years)'].fillna(df['Experience (Years)'].median())
df['Performance Rating'] = df['Performance Rating'].fillna(df['Performance Rating'].median())
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Replacing infinite values with NaN values and replacing NaN values with mean
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df['Salary (INR)'] = df['Salary (INR)'].fillna(df['Salary (INR)'].mean())
df
```

```python
# Handling negative values in Salary column
```

```python
df['Salary (INR)'] = np.where(df['Salary (INR)'] < 0, df['Salary (INR)'].median(), df['Salary (INR)'])
```

```python
df
```

```python
# Converting salary column to int from float
df['Salary (INR)'] = df['Salary (INR)'].round().astype('Int64') 
```

```python
# Handling outliers in salary column although the dataset is extremely right-skewed
```

```python
salary_mean = df['Salary (INR)'].mean()
salary_std = df['Salary (INR)'].std()
lower_bound = salary_mean - (3 * salary_std)
upper_bound = salary_mean + (2 * salary_std)
```

```python
df = df[(df['Salary (INR)'] >= lower_bound) & (df['Salary (INR)'] <= upper_bound)]
```

```python
df
```

```python
# Tracks how fast the values rise
df['Salary (INR)'].quantile([0.1, 0.25, 0.5, 0.75, 0.9])
```

## Data Analysis

#### Correlation between Salary and Experience

```python
df[['Salary (INR)', 'Experience (Years)']].corr()
```

```python
# NOTE: There is a moderately high correlation between the two entities.
```

```python
df['Exp_Bucket'] = pd.cut(df['Experience (Years)'], bins = [0, 5, 10, 20, 30, 40])

print("Average salary by Exerience Bucket")
df.groupby('Exp_Bucket')['Salary (INR)'].mean()
```

```python
# NOTE: The experience bucket of 10-20 has the highest average salary while 0-5 has the lowest.
```

#### Department-wise salaries

```python
dept_salary = df.groupby('Department')['Salary (INR)']
```

```python
dept_salary.agg(avg_salary = 'median', min_salary = 'min', max_salary = 'max', count = 'count').sort_values(by='avg_salary', ascending=False)
```

#### Performance Rating vs Salary

```python
df[['Performance Rating', 'Salary (INR)']].corr()
```

```python
# NOTE: There is a really weak correlation between Performace rating and salary so we cannot establish any conclusion using this.
```

#### City-wise salary variation within departments

```python
df.pivot_table(values='Salary (INR)', index='Department', columns='City', aggfunc='median')
```

```python
# NOTE: From this table, we can conclude that Banglore and Kochi pay the same in HR dept. while Mumbai pays a lot higher. Also, Banglore pays the most to IT dept. 
```

#### Average performance rating of each department

```python
df.groupby('Department')['Performance Rating'].mean().sort_values(ascending=False)
```

```python
# NOTE: We can see that the Operations dept. has the highest performance rating while Marketing has the lowest.
```

#### Age Bucket Analysis

```python
df[['Age', 'Performance Rating']].corr()
```

```python
# NOTE: We have a moderate correlation between the two entities.
```

```python
df['Age_Group'] = pd.cut(df['Age'], bins = [20, 30, 40 ,50, 60])
```

```python
df.groupby('Age_Group')['Performance Rating'].mean().sort_values(ascending=False)
```

```python
# NOTE: We can conclude that the age bucket of 40-50 has the highest performance rating.
```
